```python

import scipy.io as sio
import os
```

```python
name = 'Landsat_separate_images_BR_R002'
mat = sio.loadmat(f'inputs/{name}.mat')
```

```python
# replace - with _ in the keys
mat_fixed = {}
for key in mat.keys():
    mat_fixed[key.replace('-', '_')] = mat[key]
# save the new mat file
sio.savemat(f'inputs/{name}_fixed.mat', mat_fixed)
```

```python
mat = sio.loadmat(f'inputs/{name}_fixed.mat')
print(mat.keys())
```

```python
# rename/mv fixed file to the original
os.rename(f'inputs/{name}_fixed.mat', f'inputs/{name}.mat')
```
